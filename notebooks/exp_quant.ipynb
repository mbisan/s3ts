```python
# plotting libraries
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns

# data analysis libraries
import pandas as pd
import numpy as np

# general imports
from pathlib import Path

# auxiliary functions
import s3ts.analysis as aux
```

```python
plt.rc('font', family='serif')
plt.rc('xtick', labelsize='medium')
plt.rc('ytick', labelsize='medium')
```

```python
df = aux.load_data(Path("../results_synced/quant/"))
df.replace(np.NaN, 0).groupby(["arch", "dataset"]).count()
df
```

```python
dfp = aux.EXP_quant_preprocessing(df)
dfp
```

```python
dfpb = aux.EXP_quant_set_baselines(dfp)
dfpb
```

```python
fig = plt.figure(figsize=[20,8])
gs = fig.add_gridspec(2, 6, left=0.03, right=0.48, bottom=0.05,
    wspace=0., hspace=0.12)

axes: dict[str, plt.Axes] = {
    "RNN_TS": fig.add_subplot(gs[0,0:2]),
    "CNN_TS": fig.add_subplot(gs[0,2:4]),
    "ResNet_TS": fig.add_subplot(gs[0,4:6]),
    "CNN_DFS": fig.add_subplot(gs[1,0:3]),
    "ResNet_DFS": fig.add_subplot(gs[1,3:6]),
}

right = ["RNN_TS", "CNN_DFS"]

colors = {
    "GunPoint": "blue",
    "Coffee": "red",
    "PowerCons": "purple",
    "Plane": "green",
    "CBF": "orange"
}

metric = "acc"

for arch in axes:
    
    ax = axes[arch]
    ax.set_ylim(-0.10, 0.15)
    ax.set_xlim(1.5,9.5)
    ax.set_xticks(np.arange(2,10))

    ax.set_ylim(-15, 15)
    ax.set_yticks([-15, -10, -5, 0, 5, 10, 15])
    ax.grid(axis="y")

    if "TS" in arch:
        ax.set_ylim(-10, 10)
        ax.set_yticks([-10, -8, -6,  -4, -2, 0, 2, 4, 6, 8, 10])

    if arch not in right:
        ax.set_yticklabels([])

    ax.annotate(arch, (0.05, 0.05), xycoords='axes fraction', va='center')

for (arch, dataset), dfg in dfpb.groupby(["arch", "dataset"]):
    axes[arch].errorbar(dfg[f"pretrain_nquant"], dfg[f"target_val_{metric}_mean"]*100, 
        #yerr=dfg[f"target_val_{metric}_std"]*100,
        capsize=10, c=colors[dataset], linewidth=3, marker="x")

fig.text(0,0.5, s= f"abs % {metric}", rotation="vertical", ha="center", va="center");
fig.text(0.25,0, s= "number of quantiles", rotation="horizontal", ha="center", va="center");
```
